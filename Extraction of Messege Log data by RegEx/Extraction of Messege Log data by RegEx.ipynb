# Prob2. Sq feet: 


In each message, find all these words along with the number preceding it: Sq feet, sq.feet, sq.ft, sq ft

    
    Mar 19, 7:53 PM - ‪+91 90297 30114‬: Vapi pardi gujrat border near dahanu  1240 sq feet rowhouse for sale arjnt vinod

In the above example it should print 1240 sq feet.
If there are multiple occurrences, separate with comma.
There may or may not be space between number and above 'square feet' terms.

```python
fil1 = open("WhatsAppChat.txt", 'r' )
text = fil1.read()
text =text.replace('\n', '  ')
text = text.replace('Mar', '\nMar')

import re

messege_list = []

messege_Regex = re.compile(r'(Mar\s\d{2}\,\s[\d]+\:\d+\s\w\w\s\-\s‪\+\d{2}\s\d{5}\s\d{5})(.*)')
complete_list=messege_Regex.findall(text)
for no in complete_list:
    messege_list.append(no[1])
    
messege_string = ','.join(messege_list)
```

```python

sq_feet_Regex = re.compile(r'[\d]+[\s]*sqft|[\d]+[\s]*sq.ft|[\d]+[\s]*sq.feet|[\d]+[\s]*sq.\s*.ft|[\d]+[\s]*sq.\s*.feet')
sq_feet_list=sq_feet_Regex.findall(messege_string)
for no in sq_feet_list:
    print(no)
```
